## **Perceptron(Binary Classification)**

**Example : Study Hours -> Pass(1)/Fail(0)**

```python
import numpy as np
```

**Step 1 : Prepare Training data**

```python
X=np.array([1,2,3,4]) #study hours
y=np.array([0,0,1,1]) #pass(1)/fail(0)
```

```python
X,y
```

**Step 2 : Initialize Parameters**

```python
w=0.0
b=0.0
learning_rate=1
epochs=5

print("Initial Weight : ",w)
print("Initial Bias : ",b)
```

**Step 3 : Define Activation Function (Step function)**

```python
def step_function(z):
  return 1 if z>=0 else 0
```

**Training Loop**

```python
for epoch in range(epochs):
  print(f"Epoch : {epoch}")
  for i in range(len(X)):
    # 1.computer weight
    z=w*X[i]+b
    # 2.apply Activation function
    y_pred=step_function(z)
    # 3.Compute error
    error=y[i]-y_pred
    # 4.Update weight
    w=w+learning_rate*error*X[i]
    # 5.Update bias
    b=b+learning_rate*error

    print(f"  Input : {X[i]},Actual value : {y[i]},Predicted value : {y_pred},Error : {error},Updated Weight : {w},Updated Bias : {b}")
```

**Final Model Parameters**

```python
print("Final Weight : ",w)
print("Final Bias : ",b)
```

**Testing the Model**

```python
test_value=1.7
z=w*test_value+b
pred=step_function(z)
print(f"For Study hours {test_value}, Prediction : {pred}")
```

**Ploting Decision Boundary**

```python
import matplotlib.pyplot as plt

plt.scatter(X[y==0], y[y==0], color='red', label="Fail (0)")
plt.scatter(X[y==1], y[y==1], color='green', label="Pass (1)")

decision_boundary = -b/w
plt.axvline(x=decision_boundary, color='blue', linestyle='--', label="Decision Boundary")
plt.xlabel("Study Hours")
plt.ylabel("Class")
plt.title("Perceptron Decision Boundary")
plt.legend()
plt.show()
```

## **Perceptron AND Gate**

```python
X=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([0,0,0,1])
```

```python
X,y
```

```python
w=np.zeros(2)
b=0.0
learning_rate=0.1
epochs=10

print("Initial Weight : ",w)
print("Initial Bias : ",b)
```

```python
for epoch in range(epochs):
  print(f"Epoch : {epoch}")
  for i in range(len(X)):
    # 1.computer weight
    z=np.dot(w,X[i])+b
    # 2.apply Activation function
    y_pred=step_function(z)
    # 3.Compute error
    error=y[i]-y_pred
    # 4.Update weight
    w=w+learning_rate*error*X[i]
    # 5.Update bias
    b=b+learning_rate*error

    print(f"  Input : {X[i]},Actual value : {y[i]},Predicted value : {y_pred},Error : {error},Updated Weight : {w},Updated Bias : {b}")
```

```python
print("Final Weight : ",w)
print("Final Bias : ",b)
```

```python
test_value=np.array([1,0])
z=np.dot(w,test_value)+b
pred=step_function(z)
print(f"Input : {test_value}, Output : {pred}")
```

```python
for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i][0], X[i][1], color='red')
    else:
        plt.scatter(X[i][0], X[i][1], color='green')
plt.scatter([], [], color='red', label="Output: 0")
plt.scatter([], [], color='green', label="Output : 1")
x_values = np.linspace(-0.5, 1.5, 100)
y_values = -(w[0]*x_values + b) / w[1]
plt.plot(x_values, y_values, label="Decision Boundary")
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("AND Gate Decision Boundary")
plt.legend()
plt.show()
```
